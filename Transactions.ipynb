## Clean Data

```python
# accounts data frame
# Remove trailing blanks
accounts_df['acct_type'] = accounts_df['acct_type'].str.strip()
accounts_df['account_active'] = accounts_df['account_active'].str.strip()
```

```python
# Keep only active accounts
filter = accounts_df['account_active'] == 'Y'
accounts_df = accounts_df[filter]
```

```python
# drop columns that we will not use 'account_active', 'acct_start_date', 'acct_end_date'
accounts_df = accounts_df.drop(['account_active', 'acct_start_date', 'acct_end_date'], axis=1)
```

```python
# keep only columns that we will use, drop all other. Drop ['tran_id', 'principal_amt', 'interest_amt' , 'new_balance', 'tran_date', 'tran_time', 'channel', 'tran_code']
transactions_clean = transactions_df[['acct_nbr', 'tran_amt']]
```

```python
transactions_clean['tran_amt_abs'] = transactions_clean['tran_amt'].abs()
```

```python
# withdrawal
# Deposit
transactions_acct_type['tran_type'] = ''
transactions_acct_type.loc[((transactions_acct_type['acct_type']== 'SV') | (transactions_acct_type['acct_type']== 'CK') )\
                            & (transactions_acct_type['tran_amt'] < 0),'tran_type'] = 'Withdrawal'
transactions_acct_type.loc[((transactions_acct_type['acct_type']== 'SV') | (transactions_acct_type['acct_type']== 'CK') )\
                            & (transactions_acct_type['tran_amt'] >= 0),'tran_type'] = 'Deposit'
transactions_acct_type.loc[(transactions_acct_type['acct_type']== 'CC') & (transactions_acct_type['tran_amt'] < 0),'tran_type'] = 'Deposit'
transactions_acct_type.loc[(transactions_acct_type['acct_type']== 'CC') & (transactions_acct_type['tran_amt'] >= 0),'tran_type'] = 'Withdrawal'
```

```python

a = trans_summary[['acct_nbr', 'tran_type', 'tran_amt', 'tran_size']].groupby(by=['acct_nbr', 'tran_type', 'tran_size']).agg({'tran_size': 'count', 'tran_amt': 'sum'})
```

```python
a.rename(columns = {'tran_size': 'count_transize'}, inplace = True)
```

```python
b = a.pivot_table(index=['acct_nbr'], columns=['tran_type', 'tran_size'], 
                    values=['count_transize', 'tran_amt'])
```

```python
b
```

```python
b.columns.set_levels(['Deposit_Large_count', 'Deposit_Medium_count', 'Deposit_Small_count',
 'Withdrawal_Large_count' 'Withdrawal_Medium_count' 'Withdrawal_Small_count',
 'Deposit_Large_amt', 'Deposit_Medium_amt', 'Deposit_Small_amt',
 'Withdrawal_Large_amt', 'Withdrawal_Medium_amt' ,'Withdrawal_Small_amt'],level=2,inplace=True)
```

```python
b.columns = ['_'.join(col) for col in b.columns]
```

```python
trans_to_join = b.reset_index()
```

```python
trans_to_join.acct_nbr.nunique()
```

```python
#join final transactions table with account tables
trans_train = pd.merge(accounts_df, trans_to_join, how='left', on='acct_nbr')
```
